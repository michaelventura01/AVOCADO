```python
# importando dataset
import pandas as pd
import matplotlib.pyplot as plt
avocados = pd.read_csv('avocado.csv')
```

### Filas y Columnas

```python
columns = avocados.columns
totalcolumns = len(columns)
totalrows = len(avocados)
print(f'TOTAL FILAS: {totalrows}, TOTAL COLUMNAS {totalcolumns}' )
print('\nCOLUMNAS')
for column in columns:
    print(f'- {column}')
```

### Precios de Aguacates

```python
min = avocados['AveragePrice'].min()
max = avocados['AveragePrice'].max()
prom = round(avocados['AveragePrice'].mean(),2)

print(f'PRECIO MINIMO: {min} | PRECIO MAXIMO: {max} | PRECIO PROMEDIO: {prom}')
```

### Graficos

```python
ax = None
colors = ['red', 'blue', 'green'] 
for region,color in zip(avocados['region'].unique()[0:3],colors):
    subset = avocados[avocados['region'] == region]
    ax = subset.plot.scatter(x='year', y='AveragePrice', label=region, color=color, ax=ax)

plt.title('Scatter Plot de PRECIO por  ANIO por region seleccionada')
plt.grid(True)
plt.show()
```
